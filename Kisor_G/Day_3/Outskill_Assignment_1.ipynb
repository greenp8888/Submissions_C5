## Basic

```python
import gradio as gr
from transformers import pipeline

sentiment_analyzer = pipeline("sentiment-analysis")

def analyze(text):
    result = sentiment_analyzer(text)[0]
    return f"{result['label']} (score: {100 * result['score']:.3f})"

ui = gr.Interface(
    fn=analyze,
    inputs=gr.Textbox(label="Enter text"),
    outputs=gr.Textbox(label="Sentiment"),
    title="Sentiment Analyzer",
    description="Enter text and get the sentiment (positive/negative)."
)

ui.launch()
```

```python
import gradio as gr
from transformers import pipeline

sentiment_analyzer = pipeline("sentiment-analysis")

def analyze(text):
    if not text.strip():
        return "⚠️ Please enter some text."
    result = sentiment_analyzer(text)[0]
    label = result["label"]
    score = f"{result['score'] * 100:.2f}%"

    emoji = "😊" if label == "POSITIVE" else "😞"
    return f"### {emoji} {label}\n**Confidence:** {score}"

with gr.Blocks(title="Sentiment Analyzer") as ui:
    gr.Markdown(
        """
        # 🧠 Sentiment Analyzer
        Type text below and get an instant sentiment classification.
        """
    )

    with gr.Row():
        with gr.Column(scale=2):
            input_box = gr.Textbox(
                label="Enter text",
                placeholder="Type a sentence...",
                lines=4
            )
            analyze_btn = gr.Button("Analyze", variant="primary")

        with gr.Column(scale=1):
            output_box = gr.Markdown(label="Sentiment")

    examples = gr.Examples(
        examples=[
            "I love how easy this tool is!",
            "This experience was terrible.",
            "I'm not sure how I feel about this.",
        ],
        inputs=input_box
    )

    analyze_btn.click(analyze, inputs=input_box, outputs=output_box)
    input_box.change(analyze, inputs=input_box, outputs=output_box)

ui.launch()
```

```python
import gradio as gr
from transformers import pipeline
import time

sentiment_analyzer = pipeline("sentiment-analysis")

def analyze(text):
    if not text.strip():
        return "<div class='empty'>Type something to analyze sentiment.</div>"

    result = sentiment_analyzer(text)[0]
    label = result["label"]
    score = f"{result['score']*100:.2f}%"

    if label == "POSITIVE":
        color = "card-positive"
        emoji = "🍀😊"
    elif label == "NEGATIVE":
        color = "card-negative"
        emoji = "🔥😞"
    else:
        color = "card-neutral"
        emoji = "✨😐"

    return f"""
    <div class='result-card {color}'>
        <div class='emoji'>{emoji}</div>
        <div class='label'>{label}</div>
        <div class='score'>Confidence: {score}</div>
    </div>
    """

css = """
/* Background Gradient */
body {
    background: linear-gradient(135deg, #3900a6, #1b003d, #000000);
    background-size: 300% 300%;
    animation: bgmove 12s infinite;
}
@keyframes bgmove {
    0% { background-position: 0% 50%; }
    50% { background-position: 100% 50%; }
    100% { background-position: 0% 50%; }
}

/* Header */
h1 {
    text-align: center;
    color: white !important;
    font-size: 2.6rem !important;
    margin-bottom: 5px;
    text-shadow: 0 0 12px #9a00ff;
}
p {
    text-align: center;
    color: #e4d7ff !important;
    margin-top: -10px;
}

/* Input box */
textarea {
    background: rgba(255,255,255,0.1) !important;
    backdrop-filter: blur(10px);
    border-radius: 15px !important;
    color: white !important;
    padding: 15px !important;
    border: 1px solid rgba(255,255,255,0.3) !important;
    font-size: 1.2rem !important;
}

/* Empty state */
.empty {
    color: #ddd;
    text-align: center;
    padding: 20px;
    opacity: 0.7;
    font-size: 1.1rem;
}

/* Result Card container */
.result-card {
    width: 380px;
    margin: 20px auto;
    padding: 25px 20px;
    border-radius: 18px;
    text-align: center;
    color: white;
    animation: popup 0.35s ease;
    box-shadow: 0 8px 25px rgba(0,0,0,0.35);
    backdrop-filter: blur(10px);
}

/* Positive gradient */
.card-positive {
    background: linear-gradient(135deg, #1affb2, #00d97a, #00995a);
}

/* Negative gradient */
.card-negative {
    background: linear-gradient(135deg, #ff6a6a, #ff3737, #cc0000);
}

/* Neutral gradient */
.card-neutral {
    background: linear-gradient(135deg, #cfcfcf, #b4b4b4, #9e9e9e);
    color: #222;
}

/* Popup animation */
@keyframes popup {
    0% { transform: scale(0.6); opacity: 0; }
    100% { transform: scale(1); opacity: 1; }
}

/* Text inside card */
.emoji {
    font-size: 3rem;
    margin-bottom: 10px;
}
.label {
    font-size: 1.8rem;
    font-weight: bold;
    margin-bottom: 6px;
}
.score {
    font-size: 1.1rem;
    opacity: 0.9;
}
"""

with gr.Blocks(css=css, title="Gradient Sentiment Analyzer") as ui:
    gr.Markdown("""
    <h1>Sentiment Analyzer</h1>
    <p>Type text and see a beautiful sentiment card appear.</p>
    """)

    text = gr.Textbox(
        placeholder="Type something...",
        lines=3,
        show_label=False,
    )

    result = gr.HTML("<div class='empty'>Type something to analyze sentiment.</div>")

    text.change(analyze, inputs=text, outputs=result)

ui.launch()
```
